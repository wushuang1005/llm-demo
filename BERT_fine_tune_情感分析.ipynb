````python
from datasets import load_dataset

dataset = load_dataset("imdb")
print(dataset)
print(dataset["train"][0])  # 看一条样本长什么样
````

## Tokenizer — 把文字变成数字
BERT是数学模型，它看不懂英文单词，只能处理数字。Tokenizer就是做这个翻译工作的。  
具体来说分三步：  
① 切词（Tokenization）  

```
"I love this movie"   
→ ["i", "love", "this", "movie"]
```

② 映射成数字（Token IDs）  


```
→ [101, 1045, 2293, 2023, 3185, 102]
```

每个词对应BERT词典里的一个固定编号，101是句子开头的特殊符号[CLS]，102是结尾[SEP]。  
③ 统一长度（Padding & Truncation）  
BERT要求每条输入长度一样，所以：  
太短的句子 → 在后面补0（padding）  
太长的句子 → 直接截断（truncation）  
我们设的是 max_length=256


```
python# 这三个参数就是在做上面的事
tokenizer(batch["text"], padding="max_length", truncation=True, max_length=256)
```




````python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

def tokenize(batch):
    return tokenizer(batch["text"], padding="max_length", truncation=True, max_length=256)

tokenized = dataset.map(tokenize, batched=True)
print(tokenized)
````

````python
# 加载模型
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained(
    "bert-base-uncased",
    num_labels=2  # 正面 / 负面
)
````

````python
# 设置训练参数
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./bert-imdb",
    num_train_epochs=2,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=32,
    eval_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    logging_steps=100,
)
````

````python
# 设置评估指标
import evaluate
import numpy as np

accuracy = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return accuracy.compute(predictions=predictions, references=labels)
````

````python
# 开始训练
from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized["train"],
    eval_dataset=tokenized["test"],
    compute_metrics=compute_metrics,
)

trainer.train()
````

````python
# 用自己的句子测试
from transformers import pipeline

classifier = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

samples = [
    "This movie was absolutely amazing, I loved every second!",
    "What a waste of time. Terrible acting and boring plot.",
    "It was okay, nothing special but not bad either."
]

for text in samples:
    result = classifier(text)
    print(f"Text: {text}")
    print(f"Result: {result}\n")
````
